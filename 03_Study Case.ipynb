````python
import mysql.connector
import pandas as pd

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Eoi679ioe",
    database="iot_data"
)

# Set chunk size and goal per big DataFrame
chunk_size = 100_000
rows_per_df = 1_000_000

query = "SELECT * FROM ignition_log"
chunks = pd.read_sql(query, conn, chunksize=chunk_size)

# Containers
big_df_list = []
all_dfs = []  # Store df_1, df_2, etc.
current_row_count = 0
df_counter = 1

# Stream chunks and group them into 1 million-row DataFrames
for i, chunk in enumerate(chunks):
    big_df_list.append(chunk)
    current_row_count += len(chunk)

    if current_row_count >= rows_per_df:
        df = pd.concat(big_df_list, ignore_index=True)
        all_dfs.append(df)
        print(f"âœ… Created df_{df_counter} with shape: {df.shape}")
        
        # Reset for next group
        big_df_list = []
        current_row_count = 0
        df_counter += 1

# Handle remaining rows if total is not multiple of 1 million
if big_df_list:
    df = pd.concat(big_df_list, ignore_index=True)
    all_dfs.append(df)
    print(f"âœ… Created df_{df_counter} with shape: {df.shape} (last partial chunk)")

conn.close()
````

# MySQL Partitioning: Visual Example for Improved Query Performance

## Problem with a Single Large Table

Imagine you have a huge table collecting sensor data every second:

```
ignition_log
--------------------------------------------
| id | t_stamp           | Current_A | ... |
|----|-------------------|-----------|-----|
|  1 | 2025-01-01 00:00  | 32252     |     |
|  2 | 2025-01-02 00:00  | 33492     |     |
|  3 | 2025-03-01 00:00  | 34492     |     |
|  4 | 2025-06-01 00:00  | 35492     |     |
```

As data grows, this table becomes slower to query, especially for specific date ranges.

---

## Solution: Partitioning by Month

Instead of one large table, we **partition the table** based on time (e.g., by month):

```
ignition_log (partitioned)
â”œâ”€â”€ Partition: Jan 2025 (p202501)
â”œâ”€â”€ Partition: Feb 2025 (p202502)
â”œâ”€â”€ Partition: Mar 2025 (p202503)
â”œâ”€â”€ ...
```

---

##  SQL Partitioning Syntax (Example)

Hereâ€™s how to create a partitioned table in MySQL:

```sql
CREATE TABLE ignition_log (
    id BIGINT AUTO_INCREMENT PRIMARY KEY,
    t_stamp DATETIME NOT NULL,
    Current_A INT,
    ...
    INDEX(t_stamp)
)
PARTITION BY RANGE (YEAR(t_stamp) * 100 + MONTH(t_stamp)) (
    PARTITION p202501 VALUES LESS THAN (202502),
    PARTITION p202502 VALUES LESS THAN (202503),
    PARTITION p202503 VALUES LESS THAN (202504)
    -- Add more partitions as needed
);
```

---

##  Optimized Query

With partitioning, this query:

```sql
SELECT * FROM ignition_log 
WHERE t_stamp BETWEEN '2025-03-01' AND '2025-03-31';
```

Will **only search** in:

```
ðŸ“‚ Partition p202503
```

 This improves **query speed and efficiency**, especially for historical or monthly reports.

---

##  Summary

| Aspect | Without Partition | With Partition |
| --- | --- | --- |
| Query Speed | Slower (entire table scan) | Faster (partition pruning) |
| Maintenance | Harder | Easier (can drop old partitions) |
| Scalability | Limited | Highly scalable |

---

##  Tips

- Always index the `partition key` (like `t_stamp`) to enable partition pruning.
  
- You can manage old data by simply dropping old partitions:
  
  ```sql
  ALTER TABLE ignition_log DROP PARTITION p202301;
  ```
  
- Useful for **time-series** data (e.g., SCADA logs, IoT sensors, energy meters, etc.)

```python
import mysql.connector
import pandas as pd
from datetime import datetime, timedelta
import time
import os

# === Config ===
HOST = "localhost"
USER = "root"
PASSWORD = "Eoi679ioe"
DATABASE = "iot_data"
TABLE_NAME = "ignition_log"
DOWNSAMPLED_TABLE = "ignition_log_daily"
EXPORT_DIR = "exported_excels"
RETENTION_DAYS = 100
WAIT_HOURS = 24

os.makedirs(EXPORT_DIR, exist_ok=True)

def connect():
    return mysql.connector.connect(
        host=HOST,
        user=USER,
        password=PASSWORD,
        database=DATABASE
    )

def export_old_data(conn):
    cutoff_date = (datetime.now() - timedelta(days=RETENTION_DAYS)).strftime("%Y-%m-%d")
    
    query = f"""
    SELECT * FROM {TABLE_NAME}
    WHERE t_stamp < '{cutoff_date}'
    """
    df = pd.read_sql(query, conn)
    
    if df.empty:
        print(f" No data older than {cutoff_date} to export.")
        return None

    export_path = os.path.join(EXPORT_DIR, f"export_until_{cutoff_date}.xlsx")
    df.to_excel(export_path, index=False)
    print(f" Exported {len(df)} rows to {export_path}")
    return df

def delete_old_data(conn, cutoff_date):
    cursor = conn.cursor()
    delete_query = f"""
    DELETE FROM {TABLE_NAME}
    WHERE t_stamp < '{cutoff_date}'
    """
    cursor.execute(delete_query)
    conn.commit()
    cursor.close()
    print(f" Deleted rows older than {cutoff_date}")

def ensure_downsample_table(conn):
    cursor = conn.cursor()
    cursor.execute(f"""
    CREATE TABLE IF NOT EXISTS {DOWNSAMPLED_TABLE} (
        date DATE PRIMARY KEY,
        PM_ID VARCHAR(50),
        Current_A_avg FLOAT,
        Current_B_avg FLOAT,
        Current_C_avg FLOAT,
        Frequency_avg FLOAT,
        Power_Factor_avg FLOAT,
        Volt_LL_Avg_avg FLOAT,
        Active_Power_Total_avg FLOAT
    )
    """)
    conn.commit()
    cursor.close()

def insert_downsampled_data(conn):
    print(" Performing daily downsampling...")

    downsample_query = f"""
    SELECT 
        DATE(t_stamp) as date,
        PM_ID,
        AVG(Current_A) as Current_A_avg,
        AVG(Current_B) as Current_B_avg,
        AVG(Current_C) as Current_C_avg,
        AVG(Frequency) as Frequency_avg,
        AVG(Power_Factor) as Power_Factor_avg,
        AVG(Volt_LL_Avg) as Volt_LL_Avg_avg,
        AVG(Active_Power_Total) as Active_Power_Total_avg
    FROM {TABLE_NAME}
    GROUP BY DATE(t_stamp), PM_ID
    """

    df = pd.read_sql(downsample_query, conn)
    if df.empty:
        print(" No data available for downsampling.")
        return

    cursor = conn.cursor()
    insert_query = f"""
    REPLACE INTO {DOWNSAMPLED_TABLE} (
        date, PM_ID, Current_A_avg, Current_B_avg, Current_C_avg,
        Frequency_avg, Power_Factor_avg, Volt_LL_Avg_avg, Active_Power_Total_avg
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """

    cursor.executemany(insert_query, df.values.tolist())
    conn.commit()
    cursor.close()
    print(f" Downsampled and inserted {len(df)} rows into `{DOWNSAMPLED_TABLE}`")

# === Main Loop (Runs forever every 24 hours) ===
while True:
    print(" Starting scheduled maintenance at", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    try:
        conn = connect()
        ensure_downsample_table(conn)

        # Step 1: Export old data
        df_exported = export_old_data(conn)

        # Step 2: Downsample daily data
        insert_downsampled_data(conn)

        # Step 3: Delete old data if exported successfully
        if df_exported is not None:
            cutoff = (datetime.now() - timedelta(days=RETENTION_DAYS)).strftime("%Y-%m-%d")
            delete_old_data(conn, cutoff)

    except Exception as e:
        print("âŒ Error:", e)

    finally:
        if conn.is_connected():
            conn.close()
            print(" Connection closed.")

    print(f" Sleeping {WAIT_HOURS} hours until next cycle...\n")
    time.sleep(WAIT_HOURS * 3600)  # Sleep 24 hours
```
